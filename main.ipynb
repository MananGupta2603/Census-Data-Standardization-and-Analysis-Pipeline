```python
import pandas as pd
```

```python
df=pd.read_excel('census_2011.xlsx')
df
```

```python
change_column_name={
  'State name': 'State/UT',
  'District name': 'District',
  'Male_Literate': 'Literate_Male',
  'Female_Literate': 'Literate_Female',
  'Rural_Households': 'Households_Rural',
  'Urban_Households': 'Households_Urban',
  'Age_Group_0_29': 'Young_and_Adult',
  'Age_Group_30_49': 'Middle_Aged',
  'Age_Group_50': 'Senior_Citizen',
  'Age not stated': 'Age_Not_Stated'
}
df.rename(columns=change_column_name,inplace=True)
df
```

```python
p=df.drop_duplicates(subset='State/UT')
new_column=[]
old_column=[]

for i in p['State/UT']:
    old_column.append(i)
    temp=i.split()
    l=[]
    for j in temp:
        if j == 'AND' or j == 'and':
            l.append(j.lower())
        else:
            l.append(j.lower().capitalize())
    new_column.append(" ".join(l))
```

```python
for k in range(len(new_column)):
    df.replace(to_replace=old_column[k],value=new_column[k],inplace=True)
df
```

```python
change_state_name=[
'Adilabad',
'Nizamabad',
'Karimnagar',
'Medak',
'Hyderabad',
'Rangareddy',
'Mahbubnagar',
'Nalgonda',
'Warangal',
'Khammam'
]
df.loc[df['District'].isin(change_state_name),'State/UT'] = 'Telangana'
df.tail(110)
```

```python
change_state_name=[
'Leh(Ladakh)',
'Kargil'
]
df.loc[df['District'].isin(change_state_name),'State/UT'] = 'Ladakh'
df
```

```python
missing_initial = (df.isnull().mean() * 100).round(4)
#-------------------------------------------------

df['Population'].fillna(df['Male'].fillna(0) + df['Female'].fillna(0),inplace=True)
df['Literate'].fillna(df['Literate_Male'].fillna(0) + df['Literate_Female'].fillna(0),inplace=True)
df['Households'].fillna(df['Households_Rural'].fillna(0) + df['Households_Urban'].fillna(0),inplace=True )
#-------------------------------------------------
After_filling = (df.isnull().mean() * 100).round(4)
#-------------------------------------------------
new_df=pd.DataFrame({
'Before_filling%': missing_initial,
'After_filling%': After_filling
})
```
